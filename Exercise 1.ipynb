##### Attempt 3
Removing the \n and showing the emails/validity in one dataframe. (New list no_n was introduced because line = line.replace("\n", "") didn't seem to work.)

```python
with open("dataset/emails.txt","r") as file1:
    newlist=file1.readlines()
    no_n = []
    for line in newlist:
        no_n.append(line.replace('\n', ''))
    ds_no_n = pd.Series(no_n, name = "emails")
    ds_no_n = pd.DataFrame(ds_no_n)
    ds_no_n["validity"] = ds_no_n["emails"].str.extract('(^[a-zA-Z0-9][a-zA-Z0-9._-]+@[a-zA-Z]+\.[a-zA-z]+[.]?[a-zA-Z]+$)')
    print(ds_no_n)
```

##### Attempt 4

Converting the text file to a dataframe straightaway and adding a validity column. 

```python
df = pd.read_csv('dataset\emails.txt', sep="\n", header=None, names=["emails", "validity"])
df["validity"] = df["emails"].str.extract('(^[a-zA-Z0-9][a-zA-Z0-9._-]+@[a-zA-Z]+\.[a-zA-z]+[.]?[a-zA-Z]+$)')
df
```
